## Human👨‍💻 and AI 🤖 Messages

````python
from langchain.agents import create_agent
from langchain_core.messages import HumanMessage
from langchain_groq import ChatGroq

""" agent = create_agent(
    model="openai:gpt-5-nano",
    system_prompt="You are a full-stack comedian"
) """

# Initialize the Groq model
llm = ChatGroq(
    model="llama-3.1-8b-instant",
    temperature=0,
    max_retries=2,
)

# Use the llm object in the agent creation
agent = create_agent(
    model=llm,
    system_prompt="You are a full-stack comedian"
)
````

````python
human_msg = HumanMessage("Hello, how are you?")

result = agent.invoke({"messages": [human_msg]})
````

````python
print(result["messages"][-1].content)
````

````python
print(type(result["messages"][-1]))
````

````python
for msg in result["messages"]:
    print(f"{msg.type}: {msg.content}\n")
````

### Altenative formats
#### Strings
There are situations where LangChain can infer the role from the context, and a simple string is enough to create a message.

````python
agent = create_agent(
    # model="openai:gpt-5-nano",
    model=llm,
    system_prompt="You are a sports poet and comedian.",  # This is a SystemMessage under the hood
)
````

````python
result = agent.invoke({"messages": "Tell me about cricket"})   # This is a HumanMessage under the hood
print(result["messages"][-1].content)
````

#### Dictionaries

````python
result = agent.invoke(
    {"messages": {"role": "user", "content": "Write a poem about bangalore traffic"}}
)
print(result["messages"][-1].content)
````

## Output Format
### messages
Let's create a tool so agent will create some tool messages.

````python
from langchain_core.tools import tool

@tool
def check_poem_lines(text: str):
    """Check if the given poem text has exactly 3 lines.

    Returns None if it's correct, otherwise an error message.
    """
    # Split the text into lines, ignoring leading/trailing spaces
    lines = [line.strip() for line in text.strip().splitlines() if line.strip()]
    print(f"checking poem, it has {len(lines)} lines:\n {text}")

    if len(lines) != 3:
        return f"Incorrect! This poem has {len(lines)} lines. A poem must have exactly 3 lines."
    return "Correct, this poem has 3 lines."
````

````python
agent = create_agent(
    #model="openai:gpt-5",
    model=llm,
    tools=[check_poem_lines],
    system_prompt="You are a sports poet who only writes comedy poems. You always ensure poems are of 3 lines (new lines)",
)
````

````python
result = agent.invoke({"messages": "Please write me a poem"})
````

````python
result["messages"][-1].content
````

````python
print(len(result["messages"]))
````

````python
for i, msg in enumerate(result["messages"]):
    msg.pretty_print()
````
